# 基础网页爬虫教程

本 notebook 演示如何使用 requests 和 BeautifulSoup 进行基础的网页数据抓取。

## 学习目标
- 发送 HTTP 请求
- 解析 HTML 内容
- 提取特定数据
- 保存数据到文件

```python
# 导入必要的库
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sys
sys.path.append('..')
from utils.helpers import get_random_headers, save_to_csv
```

## 1. 发送 HTTP 请求

```python
# 目标网站
url = "https://quotes.toscrape.com/"

# 发送请求
response = requests.get(url, headers=get_random_headers())
print(f"状态码: {response.status_code}")
print(f"响应长度: {len(response.text)} 字符")
```

## 2. 解析 HTML 内容

```python
# 使用 BeautifulSoup 解析
soup = BeautifulSoup(response.text, 'lxml')

# 查看页面标题
print(f"页面标题: {soup.title.string}")
```

## 3. 提取名言数据

```python
# 提取所有名言
quotes = []

for quote in soup.find_all('div', class_='quote'):
    text = quote.find('span', class_='text').text
    author = quote.find('small', class_='author').text
    tags = [tag.text for tag in quote.find_all('a', class_='tag')]
    
    quotes.append({
        '名言': text,
        '作者': author,
        '标签': ', '.join(tags)
    })

print(f"共提取 {len(quotes)} 条名言")
```

## 4. 查看数据

```python
# 转换为 DataFrame
df = pd.DataFrame(quotes)
df.head()
```

## 5. 保存数据

```python
# 保存到 CSV
save_to_csv(quotes, '../data/quotes.csv')
```
