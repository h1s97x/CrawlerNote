# API 数据获取

本 notebook 演示如何通过 API 获取结构化数据。

## 学习目标
- 调用 REST API
- 处理 JSON 数据
- API 认证和参数传递

```python
import requests
import pandas as pd
import json
```

## 1. 简单的 GET 请求

```python
# 使用公开 API（JSONPlaceholder）
url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)
data = response.json()

print(f"获取了 {len(data)} 条记录")
print(f"\n第一条记录:")
print(json.dumps(data[0], indent=2, ensure_ascii=False))
```

## 2. 带参数的 API 请求

```python
# 使用查询参数
params = {
    'userId': 1
}

response = requests.get(url, params=params)
user_posts = response.json()

print(f"用户 1 共有 {len(user_posts)} 篇文章")
```

## 3. 转换为 DataFrame

```python
df = pd.DataFrame(data)
df.head()
```

## 4. 数据统计分析

```python
# 按用户统计文章数
posts_per_user = df.groupby('userId').size()
print("每个用户的文章数:")
print(posts_per_user)
```

## 5. 可视化

```python
import matplotlib.pyplot as plt

posts_per_user.plot(kind='bar', figsize=(10, 6))
plt.title('每个用户的文章数量')
plt.xlabel('用户 ID')
plt.ylabel('文章数')
plt.tight_layout()
plt.show()
```
