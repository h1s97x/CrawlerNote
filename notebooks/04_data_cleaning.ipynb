# 数据清洗与处理

本 notebook 演示如何清洗和处理爬取的数据。

## 学习目标
- 处理缺失值
- 数据类型转换
- 文本清洗
- 数据去重

```python
import pandas as pd
import re
import numpy as np
```

## 1. 加载数据

```python
# 创建示例数据
data = {
    '标题': ['  Python 教程  ', 'Java 入门', None, 'Python 教程', 'C++ 基础'],
    '价格': ['¥99.00', '199', 'N/A', '99', '299.5'],
    '评分': ['4.5分', '4.8', '5.0', '4.5', None],
    '日期': ['2024-01-15', '2024/02/20', '2024.03.10', '2024-01-15', '2024-04-05']
}

df = pd.DataFrame(data)
print("原始数据:")
df
```

## 2. 处理缺失值

```python
# 查看缺失值
print("缺失值统计:")
print(df.isnull().sum())

# 填充缺失值
df['标题'].fillna('未知', inplace=True)
df['评分'].fillna('0', inplace=True)
```

## 3. 文本清洗

```python
# 去除空格
df['标题'] = df['标题'].str.strip()

# 清洗价格数据
def clean_price(price):
    if pd.isna(price) or price == 'N/A':
        return np.nan
    # 移除货币符号和非数字字符
    price_str = re.sub(r'[^0-9.]', '', str(price))
    return float(price_str) if price_str else np.nan

df['价格'] = df['价格'].apply(clean_price)

# 清洗评分数据
df['评分'] = df['评分'].str.replace('分', '').astype(float)

print("清洗后的数据:")
df
```

## 4. 数据类型转换

```python
# 统一日期格式
df['日期'] = df['日期'].str.replace('/', '-').str.replace('.', '-')
df['日期'] = pd.to_datetime(df['日期'])

print("数据类型:")
print(df.dtypes)
```

## 5. 数据去重

```python
# 删除重复行
print(f"去重前: {len(df)} 行")
df_clean = df.drop_duplicates(subset=['标题', '日期'])
print(f"去重后: {len(df_clean)} 行")

df_clean
```

## 6. 数据验证

```python
# 检查数据范围
print("价格统计:")
print(df_clean['价格'].describe())

print("\n评分统计:")
print(df_clean['评分'].describe())
```

## 7. 保存清洗后的数据

```python
df_clean.to_csv('../data/cleaned_data.csv', index=False, encoding='utf-8-sig')
print("数据已保存")
```
