## 1. 文本文件读写

```python
# 写入文本文件
content = """这是第一行
这是第二行
这是第三行"""

# 方法1: 基础写入
with open('test.txt', 'w', encoding='utf-8') as f:
    f.write(content)

print("✓ 文件写入成功")

# 读取文本文件
with open('test.txt', 'r', encoding='utf-8') as f:
    data = f.read()
    print("\n读取的内容:")
    print(data)

# 按行读取
print("\n按行读取:")
with open('test.txt', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f, 1):
        print(f"第{i}行: {line.strip()}")
```

```python
# 追加写入
with open('test.txt', 'a', encoding='utf-8') as f:
    f.write("\n这是追加的内容")

print("✓ 追加写入成功")

# 读取所有行到列表
with open('test.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    print(f"\n共有 {len(lines)} 行")
    print("最后一行:", lines[-1].strip())
```

## 2. CSV 文件处理

```python
import csv

# 写入 CSV 文件
headers = ['书名', '价格', '作者']
books = [
    ('流浪地球', 18, '刘慈欣'),
    ('三体', 23, '刘慈欣'),
    ('时间简史', 35, '霍金')
]

with open('books.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    writer.writerows(books)

print("✓ CSV 文件写入成功")

# 读取 CSV 文件
print("\n读取 CSV:")
with open('books.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
import csv

# 使用字典写入 CSV
data = [
    {'name': '流浪地球', 'price': 18, 'author': '刘慈欣'},
    {'name': '三体', 'price': 23, 'author': '刘慈欣'},
    {'name': '时间简史', 'price': 35, 'author': '霍金'}
]

with open('books_dict.csv', 'w', encoding='utf-8', newline='') as f:
    fieldnames = ['name', 'price', 'author']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    
    writer.writeheader()
    writer.writerows(data)

print("✓ 字典格式 CSV 写入成功")

# 读取为字典
print("\n读取为字典:")
with open('books_dict.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['name']}: ¥{row['price']} - {row['author']}")
```

## 3. JSON 数据处理

```python
import json

# Python 对象转 JSON
data = {
    'movies': [
        {'name': '流浪地球', 'score': 8.5, 'year': 2019},
        {'name': '三体', 'score': 9.0, 'year': 2023}
    ],
    'total': 2
}

# 写入 JSON 文件
with open('movies.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("✓ JSON 文件写入成功")

# 读取 JSON 文件
with open('movies.json', 'r', encoding='utf-8') as f:
    loaded_data = json.load(f)

print("\n读取的 JSON 数据:")
print(json.dumps(loaded_data, ensure_ascii=False, indent=2))
```

```python
import json

# JSON 字符串操作
json_str = '{"name": "Python", "version": 3.8}'

# 解析 JSON 字符串
obj = json.loads(json_str)
print("解析结果:", obj)
print("类型:", type(obj))

# 对象转 JSON 字符串
data = {'language': 'Python', 'features': ['简单', '强大', '优雅']}
json_output = json.dumps(data, ensure_ascii=False, indent=2)
print("\nJSON 字符串:")
print(json_output)
```

## 4. XML 数据处理

```python
import xml.etree.ElementTree as ET

# 创建 XML
root = ET.Element('books')

book1 = ET.SubElement(root, 'book', id='1')
ET.SubElement(book1, 'name').text = '流浪地球'
ET.SubElement(book1, 'author').text = '刘慈欣'
ET.SubElement(book1, 'price').text = '18'

book2 = ET.SubElement(root, 'book', id='2')
ET.SubElement(book2, 'name').text = '三体'
ET.SubElement(book2, 'author').text = '刘慈欣'
ET.SubElement(book2, 'price').text = '23'

# 写入 XML 文件
tree = ET.ElementTree(root)
tree.write('books.xml', encoding='utf-8', xml_declaration=True)

print("✓ XML 文件创建成功")
```

```python
import xml.etree.ElementTree as ET

# 读取 XML 文件
tree = ET.parse('books.xml')
root = tree.getroot()

print("XML 内容:")
print(f"根元素: {root.tag}")
print("\n书籍列表:")

for book in root.findall('book'):
    book_id = book.get('id')
    name = book.find('name').text
    author = book.find('author').text
    price = book.find('price').text
    print(f"ID: {book_id}, 书名: {name}, 作者: {author}, 价格: ¥{price}")
```

## 5. 日志记录

```python
import logging
from datetime import datetime

# 配置日志
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('crawler.log', encoding='utf-8'),
        logging.StreamHandler()
    ]
)

logger = logging.getLogger(__name__)

# 使用日志
logger.info("爬虫开始运行")
logger.debug("这是调试信息")
logger.warning("这是警告信息")
logger.error("这是错误信息")

print("\n✓ 日志已记录到 crawler.log")
```

```python
import logging

# 自定义日志类
class CrawlerLogger:
    """爬虫日志工具"""
    
    def __init__(self, name='crawler', log_file='crawler.log'):
        self.logger = logging.getLogger(name)
        self.logger.setLevel(logging.DEBUG)
        
        # 文件处理器
        fh = logging.FileHandler(log_file, encoding='utf-8')
        fh.setLevel(logging.DEBUG)
        
        # 控制台处理器
        ch = logging.StreamHandler()
        ch.setLevel(logging.INFO)
        
        # 格式化
        formatter = logging.Formatter(
            '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
        )
        fh.setFormatter(formatter)
        ch.setFormatter(formatter)
        
        self.logger.addHandler(fh)
        self.logger.addHandler(ch)
    
    def info(self, msg):
        self.logger.info(msg)
    
    def error(self, msg):
        self.logger.error(msg)
    
    def warning(self, msg):
        self.logger.warning(msg)
    
    def debug(self, msg):
        self.logger.debug(msg)

# 使用示例
log = CrawlerLogger('my_crawler')
log.info("开始爬取数据")
log.warning("检测到反爬虫机制")
log.error("连接超时")

print("\n✓ 自定义日志类使用完成")
```

## 6. 综合示例：数据管道

```python
import json
import csv
import logging
from datetime import datetime

class DataPipeline:
    """数据处理管道"""
    
    def __init__(self, name='pipeline'):
        self.name = name
        self.logger = self._setup_logger()
    
    def _setup_logger(self):
        logger = logging.getLogger(self.name)
        logger.setLevel(logging.INFO)
        
        handler = logging.StreamHandler()
        formatter = logging.Formatter('%(asctime)s - %(message)s')
        handler.setFormatter(formatter)
        logger.addHandler(handler)
        
        return logger
    
    def save_to_json(self, data, filename):
        """保存为 JSON"""
        try:
            with open(filename, 'w', encoding='utf-8') as f:
                json.dump(data, f, ensure_ascii=False, indent=2)
            self.logger.info(f"✓ 保存到 JSON: {filename}")
            return True
        except Exception as e:
            self.logger.error(f"✗ JSON 保存失败: {e}")
            return False
    
    def save_to_csv(self, data, filename, headers):
        """保存为 CSV"""
        try:
            with open(filename, 'w', encoding='utf-8', newline='') as f:
                writer = csv.DictWriter(f, fieldnames=headers)
                writer.writeheader()
                writer.writerows(data)
            self.logger.info(f"✓ 保存到 CSV: {filename}")
            return True
        except Exception as e:
            self.logger.error(f"✗ CSV 保存失败: {e}")
            return False
    
    def save_to_txt(self, data, filename):
        """保存为文本"""
        try:
            with open(filename, 'w', encoding='utf-8') as f:
                f.write(f"生成时间: {datetime.now()}\n")
                f.write("="*60 + "\n")
                for item in data:
                    f.write(f"{item}\n")
            self.logger.info(f"✓ 保存到 TXT: {filename}")
            return True
        except Exception as e:
            self.logger.error(f"✗ TXT 保存失败: {e}")
            return False

# 使用示例
pipeline = DataPipeline('data_pipeline')

# 示例数据
movies = [
    {'name': '流浪地球', 'score': 8.5, 'year': 2019},
    {'name': '三体', 'score': 9.0, 'year': 2023}
]

# 保存为不同格式
pipeline.save_to_json(movies, 'output.json')
pipeline.save_to_csv(movies, 'output.csv', ['name', 'score', 'year'])
pipeline.save_to_txt([f"{m['name']} - {m['score']}" for m in movies], 'output.txt')

print("\n✓ 数据管道演示完成")
```
